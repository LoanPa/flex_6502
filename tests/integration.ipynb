```python
import sys
import os
sys.path.insert(0, os.path.abspath('..'))
import py4hw
from description import Alu

hw = py4hw.HWSystem()

#from description.alu import Alu
#from description.datapath import Datapath
a = hw.wire("a", 8)
b = hw.wire("b", 8)
sel_op = hw.wire("sel", 3)
result = hw.wire("result", 8)

carry_in = hw.wire("carry_in", 1)
zero = hw.wire("zero", 1)
negative = hw.wire("negative", 1)
carry_out = hw.wire("carry_out", 1)
overflow = hw.wire("overflow", 1)

alu = Alu(hw, "alu", a, b, sel_op, carry_in, result, zero, negative, carry_out, overflow)

print('Alu created')
```

```python
from description import Datapath
#result
pc_inc = hw.wire("pc_inc", 1)
mbr_enable = hw.wire("mbr_enable", 1)
enable_sel = hw.wire("enable_sel", 4)
write = hw.wire("write", 1)
reg_sel_a = hw.wire("reg_sel_a", 4)
reg_sel_b = hw.wire("reg_sel_b", 4)
mbr_src_sel = hw.wire("mbr_src_sel", 1)
clear_marh = hw.wire("clear_marh", 1)
mm_data_in = hw.wire("mm_data_in", 8)
#a
#b
mm_data_out = hw.wire("mm_data_out", 8)
mm_addr_src = hw.wire("mm_addr_src", 1)
mm_addr_out = hw.wire("mm_addr_out", 16)
ir_data_out = hw.wire("ir_data_out", 8)
# negative
# overflow
# zero
# carry
flag_in_sel = hw.wire("flag_in_sel", 1)

n_enable = hw.wire("n_enable", 1)
v_enable = hw.wire("v_enable", 1)
b_enable = hw.wire("b_enable", 1)
i_enable = hw.wire("i_enable", 1)
z_enable = hw.wire("z_enable", 1)
c_enable = hw.wire("c_enable", 1)


datapath = Datapath(hw, "datapath",
                    result,
                    pc_inc, mbr_enable,
                    enable_sel, write,
                    reg_sel_a, reg_sel_b,
                    a, b,
                    mbr_src_sel, clear_marh,
                    mm_data_in, mm_data_out,
                    mm_addr_src, mm_addr_out,
                    ir_data_out,
                    negative, overflow, zero, carry_out, carry_in,
                    flag_in_sel,
                    n_enable, v_enable, b_enable, i_enable, z_enable, c_enable)

print('Datapath created')
print(type(datapath.outPorts[2].wire))
```
